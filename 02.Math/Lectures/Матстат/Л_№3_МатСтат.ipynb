## Проверка статистических гипотез

До сегодняшнего дня мы жили в мире, в котором предполагается единственно верная статистическая модель. Расширим возможности статистики предложив аппарат для сравнения правдоподобности предлагаемых моделей, а конкретно для возможности выбора одной модели в приоритет другой.

Пусть выбран некоторый класс распределений $\mathcal P$. В нём выберем подклассы $\mathcal P_0$ и $\mathcal P_1$ так, чтобы $\mathcal P_0 \cap \mathcal P_1 = \emptyset$.

##### Определение
Пусть дана выборка $\mathbf X^n$ из неизвестного распределения. Гипотеза $H$ - это предположение о том, что истинное распределение выборки лежит в некотором классе распределений $\mathcal P$.

Проверить статистическую гипотезу - значит принять решение об отвержении гипотезы $H_0$ против альтернативы $H_1$.

$\textit{Замечание.}$ Мы не принимаем гипотезу, мы либо отвергаем, либо не отвергаем её. Это связано с тем, что выбранный для нулевой гипотезы класс распределений в случае принятия мог бы быть слишком широким для истинного распределения выборки. Удобнее брать достаточно широкую гипотезу и пытаться её отвергнуть, а при необходимости затем можно ставить более узкую гипотезу.

##### Определение
Пусть задана гипотеза $H_0$ против альтернативы $H_1$. Критерием проверки гипотезы называется следующее правило: гипотеза $H_0$ отвергается в пользу альтернативы в случае $X^n \in S$. Подмножество пространства значений выборки $S \subset \mathcal X$ -  некоторое множество, называемое критическим.

Мы и будем заниматься построением таких правил с помощью нахождения критических множеств и анализировать их свойства.

Простой пример. В модели рапсределения Бернулли с вероятностью $p$ сравним гипотезу $H_0$: $p = 1$ против альтернативы $H_1: p < 1$. Будем отвергать гипотезу $H_0$, если

$\mathbf X^n \in S = \{\mathbf X^n: X_{(1)} = 0\}$.

Понятно, что такая гипотеза является "хорошей", так как мы можем отвергнуть её только тогда, когда в выборке есть ноль, а значит, вероятность нуля не нулевая. В общем случае о "хорошести" гипотез можно говорить только в вероятностном смысле.

$\textit{Замечание.}$ Иногда мы будем опускать громоздкую запись $\mathbf X^n \in S = \{\mathbf X^n: \}$ и будем записывать критические множества в виде $T(\mathbf X^n) > t$, где $T$ - некоторая статистика выборки, $t$ - константа.

#### Ошибки

Так как события, которые мы исследуем, носят случайный характер, понятно, что идеальных критериев проверки не существует и всегда может произойти ошибка. Однако, ошибки делятся на два вида.

- ошибка первого рода: нулевая гипотеза отвергнута при условии того, что она верна
- ошибка второго рода: нулевая гипотеза не отвергнута при условии того, что она не верна

Конечно, в принятом нами алгоритме ошибка первого рода более серьёзна, так как если гипотеза отвергнута, то мы к ней уже не возвращаемся и можем навсегда потерять верный класс распределений. В случае если мы не отвергли неверную гипотезу мы можем построить другую, более узкую, и попробовать снова.

Поэтому свойства "хорошего" критерия таковы: вероятность ошибки первого рода ограничена, а вероятность ошибки второго рода достаточно мала.

В примере выше вероятность ошибки первого рода равна 0: 
$\mathbb P\left(X_{(1)} = 0 | p = 1\right) = 0$.

Вероятность ошибки второго рода
$\mathbb P\left(X_{(1)} = 1 | p < 1\right) = p^n$.

Для обобщения качества критериев введём определения.

##### Определение
Функцией мощности, принимающей распределение и критерий, называется действительнозначная функция $\beta(\mathbb P, S) = \mathbb P(\mathbf X^n \in S)$.

##### Определение
Критерий имеет уровень значимости $\alpha$, если для любого $\mathbb P \in \mathcal P_0$ выполнено

$\beta(\mathbb P, S) \leq \alpha$.

Таким образом, если критерий имеет уровень значимости $\alpha$, то вероятность ошибки первого рода ограничена сверху $\alpha$.

##### Определение
Размером критерия называется его минимальный уровень значимости

$\gamma = \sup_{\mathcal P_0} \beta(\mathbb P, S)$.

Размер критерия - это максимальная по классу $\mathcal P_0$ вероятность ошибки первого рода

#### p-value

На просторах интернета и в различных статьях можно найти такое определение p-value:

Пусть задан критерий вида $T(X) > t$. p-value - это вероятность появления "экстремальной" выборки при условии верности гипотезы $H_0$:

$\mathbb P(T(X) > t | H_0)$.

У такого определения есть два недостатка. 
- Во-первых, это практически вероятность ошибки первого рода, за исключением того, что p-value скорее используется как мера "экстремальности" выборки а не как характеристика критерия
- Во-вторых, такое значение p-value часто использую неправильно (даже статья на вики есть https://en.wikipedia.org/wiki/Misuse_of_p-values) в качестве табличных значений для принятия/отвержения гипотез. Это ошибка, так как $P(H|X) \neq P(X|H)$.

Мы будем использовать другое, более информативное определение

##### Определение
p-value - это наименьший уровень значимости, при котором можно отвергнуть гипотезу.

Упражнение. Почему такое определение логично?

Упражнение. Подумать, почему верно утверждение: чем больше p-value, тем вероятнее, что $H_0$ надо отвергнуть.

#### Построение критериев с помощью доверительных интервалов

В модели параметрического семейства распределений $\mathbb P_{\theta}$ с помощью выборки $\mathbf X^n$ построим доверительный интервал $(L(\mathbf X^n), U(\mathbf X^n))$ уровня доверия $1 - \alpha$. Это означает, что

$\forall \theta: \mathbb P_{\theta}\left(L(\mathbf X^n) \leq \theta \leq U(\mathbf X^n)\right) \geq 1 - \alpha$.

Для проверки гипотезы $H_0: \theta = \theta_0$ против альтернативы $H_1: \theta \neq \theta_0$ используем критерий

$S = \{\mathbf X^n: \theta_0 \notin [L(\mathbf X^n), U(\mathbf X^n)]\}$.

Такой критерий имеет уровень значимости $\alpha$, так как

$\mathbb P_{\theta_0}\left(\theta_0 \notin [L(\mathbf X^n), U(\mathbf X^n)]\right) = 1 - \mathbb P_{\theta}\left(L(\mathbf X^n) \leq \theta \leq U(\mathbf X^n)\right) \leq \alpha.$

Таким образом, проверка простой гипотезы эквивалентна проверке попадания гипотетического значения параметра в доверительный интервал.

Пример. В модели нормального распределения с дисперсией 1 для проверки гипотезы $H_0: \mu = 0$ против альтернативы $H_1: \mu \neq 0$ критерий уровня значимости $0.05$:

$S = \left\{\mathbf X^n: 0 \notin \left[\overline{\mathbf X^n} - \frac{z_{0.975}}{\sqrt{n}}, \overline{\mathbf X^n} - \frac{z_{0.025}}{\sqrt{n}}\right]\right\}$.

$\textit{Теорема.}$ Если $L, U$ - состоятельные оценки параметра $\theta$, то вероятность ошибок первого и второго рода для такого критерия стремится к нулю при $n \to \infty$.

#### Построение критериев с помощью квантилей

Построение критериев, в отличие от построения доверительных интервалов, наоборот требует поиска области, в которой искомый параметр не лежит. Таким образом можно упростить критерии.

Вспомнив метод центральной статистики, в необходимой модели найдём функцию $G(\theta, \mathbf X^n)$ распределение которой $\mathbb P_G$ не зависит от $\theta$. Тогда для любого $\theta$ и квантиля $z$ распределения $\mathbb P_G$

$\mathbb P(G(\theta, \mathbf X^n) > z_{1 - \alpha}) \leq \alpha$.

Тогда если $G$ - удобная функция, можно построить критерий вида $T(\mathbf X^n) > t$ необходимого уровня значимости.

Упражнение. В чём в данном случае смысл p-value?

Пример. В той же модели, что и предыдущий пример, для проверки гипотезы $H_0: \mu = 0$ критерий

$S = \left\{\textbf X^n: \sqrt{n}\overline{\textbf X^n} > z_{1 - \alpha}\right\}$

имеет уровень значимости $\alpha$.

### Практическая часть

Пусть $\mathbf X^n$ - выборка из распределения $\mathcal N(\mu, 1)$. Постройте критерий для проверки гипотезы $H_0: \mu < 0$ против альтернативы $H_1: \mu \geq 0$ вида $\overline{\mathbf X^n} > c$.

- Найдите какой-либо уровень значимости критерия
- Выразите его размер через $c$
- Подберите $c$ так, чтобы размер критерия был равен $\gamma=0.05$
- Сгенерируйте выборку для $\mu = -1$. Отвергается ли гипотеза?
- Проверьте найденный размер критерия экспериментально. Для этого для различных $\mu < 0$ стоит проверить, что доля ошибок первого рода не сильно превосходит $\alpha$. Для значений $\mu \in [-1, 0]$ сгенерируйте $M = 100$ выборок и найдите долю выборок, на которой гипотеза отвергается. 
- постройте график зависимости доли ошибок первого рода от значения $\mu$. Как он меняется при увеличении $M$? Почему?
- постройте график зависимости доли ошибок второго рода от значения $\mu$ в интервале $[0, 1]$.

1. $\forall \epsilon > 0: \alpha = 1 - \Phi[\sqrt{n}(c - \mu)] + \epsilon$

2. $\gamma = 1 - \Phi(c\sqrt{n})$

3. $c = \frac{z_{1-\gamma}}{\sqrt{n}}$

```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
z = sts.norm().ppf(q=0.95)
z
```

```python
def criteria(sample, constant):
    return np.mean(sample, axis=-1) > constant
```

```python
N = 100
alpha = 0.05
```

```python
c = sts.norm(0, 1).ppf(q=1-alpha) / np.sqrt(N)
```

```python
check_sample = sts.norm(-1, 1).rvs(N)
print(f"Hypothesis H0 {'' if criteria(check_sample, c) else 'not'} rejected")
```

```python
mus = np.linspace(-1, 0, 101)
mus
```

```python
M = 10000
estimates = []

for mu in mus:
    samples = sts.norm(mu, 1).rvs(size=(M, N))
    rejects = criteria(samples, c)
    estimates.append(rejects.sum() / M)
```

```python
plt.figure(figsize=(12, 10))
plt.grid()
plt.plot(mus, [alpha]*len(mus), label='size')
plt.plot(mus, estimates, label=r'$\hat{p_1}$')

beta = 1 - sts.norm().cdf(sts.norm().ppf(q=0.95) - np.sqrt(N) * mus)
plt.plot(mus, beta, label=r'$\beta(\mu)$')


plt.xlabel(r'$\mu$')
plt.legend()
```
