### Exercise 1.2

Print all sublists of the list `nucleotides`. Note: [a, c] is a sublist for [a, b, c].

Sample input: `[0, 1]`

Sample output:
*   []
*   [0]
*   [1]
*   [0, 1]


````python
nucleotides = ['A', 'T', 'G', 'C']
````

````python
# your code here
sublists=[[]]
for i in range(len(nucleotides)+1):
  for j in range(len(nucleotides)+1):
    sublst=nucleotides[i:j]
    if len(sublst)!=0:
        sublists.append(sublst)

for i in sublists:
    print(i)
len(sublists)
````

````python
# доработанный вариант
nucleotides = ['A', 'T', 'G', 'C']
sublists=[[]]
for i in range(len(nucleotides)+1):
  for k in range(i,len(nucleotides)):
    for j in range(k,len(nucleotides)):
      sublist=(nucleotides[i:k])
      sublist.append(nucleotides[j])
      if sublist is not None and sublist not in sublists:
        sublists.append(sublist)

for i in sublists:
    print(i)
len(sublists)
````

````python
# доработанный вариант 2
nucleotides = ['A', 'T', 'G', 'C']
sublists=[[],nucleotides]
for i in range(len(nucleotides)+1):
  for k in range(i,len(nucleotides)):
    for j in range(k,len(nucleotides)):
        sublist=[]
        sublist.append(nucleotides[i])
        if nucleotides[k] not in sublist: 
            sublist.append(nucleotides[k])
            if nucleotides[j] not in sublist:
                sublist.append(nucleotides[j])

        if sublist is not None and sublist not in sublists:
            sublists.append(sublist)

for i in sublists:
    print(i)
len(sublists)
````

### Exercise 1.4

Create 'decoded' list from 'encoded'

Example: 

```encode_list = [0, [1, 2], 4, [1, 0, 3]]```

```decode_list = [0, 1, 2, 4, 1, 0, 3]```



````python
# your code here
encode_list = [0, [1, 2], 4, [1, 0, 3]]
decode_list=[]
for el in encode_list:
  if type(el) is list:
    for subel in el:
      decode_list.append(subel)
  else:
    decode_list.append(el)
print(decode_list)
````

### Exercise 2.2

Calculate the sum and average for each tuple in the given list. Return a list of `(sum, mean)` tuples

````python
list_of_tuples = [(1, 2, 6), (2, 3, -6), (3, 4), (2, 2, 2, 2)]
# your code here

result_list=[]

for sub_tuple in list_of_tuples:
  sub_sum=0
  for i in sub_tuple:
    sub_sum+=i
  sub_mean=sub_sum/len(sub_tuple)
  tuple_res=sub_sum,sub_mean
  result_list.append(tuple_res)
print(result_list)
````

### Exercise 2.3

Given a list (create one for tests) with strings "name number". Return a tuple of tuples ("name", number) sorted in decreasing order by number. For same numbers sort alphabetically.

````python
# your code here
nn=["zigmund 1","boris 3","alina 2"]
tup=tuple()
for i in nn:
  sublist=i.split()
  sublist[1]=int(sublist[1])
  subtuple=tuple(sublist)
  tup+=(subtuple,)
alphabetic_ord=tuple(sorted(tup))
dec_ord_numb=tuple(sorted(tup,key=lambda x:x[1],reverse=True))

print(alphabetic_ord)
print(dec_ord_numb)
````

````python
# исправленный код
import operator
nn=["zigmund 1","boris 3","boris 2","alina 2"]
lst=[]
for i in nn:
  sublist=i.split()
  sublist[1]=int(sublist[1])
  subtuple=tuple(sublist)
  lst.append(subtuple)
print(lst)
a = sorted(lst,key=operator.itemgetter(0)) # сортировка по имени
#print(a)
b = sorted(a,key=operator.itemgetter(1),reverse=True) #сортировка (отсортированного по имени) по номеру с реверсией
print(b)
````

### Exercise 3.1

Write a program to merge several dictionaries into one. If the keys match, then leave the last encountered element.

````python
dict_1 = {'key1': 10, 'key2': 20}
dict_2 = {'key3': 30, 'key4': 40}
dict_3 = {'key1': 50, 'key5': 60}

# your code here
dict_4={}
lst=[dict_1,dict_2,dict_3]
for i in lst:
  dict_4.update(i)
print(dict_4)
````

### Exercise 3.2

Write a program to merge several dictionaries into one. If the keys match, then leave the first encountered element.

````python
dict_1 = {'key1': 10, 'key2': 20}
dict_2 = {'key3': 30, 'key4': 40}
dict_3 = {'key1': 50, 'key5': 60}

# your code here
dict_4={}
dict_4.update(dict_1)
lst=[dict_2,dict_3]
for dicts in lst:
  for keys,values in dicts.items():
    if keys not in dict_4:
      dict_4[keys]=values
print(dict_4)
````

### Exercise 3.4

Create dictionary `{"word": occurrences}` from `python_words`. Print the most common word and the number of occurrences. Don't use `collections` 

````python
# your code here
dict_5={}
for i in python_words:
  if i not in dict_5:
    dict_5[i]=1
  else:
    dict_5[i]=dict_5[i]+1

maxi=-1
for keys,values in dict_5.items():
  if values>maxi:
    maxi=values
    mostcommon=keys

print(mostcommon,maxi)
print(dict_5)
````

Generate two lists like in the previous ex. Print all numbers that meet only once.
